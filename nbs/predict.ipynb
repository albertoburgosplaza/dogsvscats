```python
from dogsvscats import config
from dogsvscats.model.model import get_model

model = get_model(model_name="resnet18_2021-01-26")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from dogsvscats.data.dataset import pil_loader 
from dogsvscats.model.predict import predict_image

SAMPLE_NUMBER = np.random.randint(1, 12501)
SAMPLE_PATH = config.TEST_DATA_PATH / f"{SAMPLE_NUMBER}.jpg"
image = pil_loader(SAMPLE_PATH)
preds = predict_image(model, SAMPLE_PATH)
label = config.CLASSES[preds]
plt.imshow(image)
plt.title(label)
plt.show()
```
